# Hierarchical clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('mall_customers.csv')
# dataset_wide = dataset.iloc[ : , 1: ].values
dataset_narrow_m = dataset.iloc[ : , -2:].values
```

## Using the dendrogam to find the optimal numbers of clusters

```python
import scipy.cluster.hierarchy as shc
dendrogram = shc.dendrogram(shc.linkage(dataset_narrow_m, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
# plt.rcParams["figure.figsize"] = (20, 10)
plt.show()
```

## Training the model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(dataset_narrow_m)
```

## Visualising the clusters

```python
plt.scatter(dataset_narrow_m[y_hc == 0, 0], dataset_narrow_m[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(dataset_narrow_m[y_hc == 1, 0], dataset_narrow_m[y_hc == 1, 1], s = 100, c = 'green', label = 'Cluster 2')
plt.scatter(dataset_narrow_m[y_hc == 2, 0], dataset_narrow_m[y_hc == 2, 1], s = 100, c = 'blue', label = 'Cluster 3')
plt.scatter(dataset_narrow_m[y_hc == 3, 0], dataset_narrow_m[y_hc == 3, 1], s = 100, c = 'yellow', label = 'Cluster 4')
plt.scatter(dataset_narrow_m[y_hc == 4, 0], dataset_narrow_m[y_hc == 4, 1], s = 100, c = 'purple', label = 'Cluster 5')
# plt.scatter(hc.[ : , 0], dataset_narrow_m[ : , 1], s = 100, c = 'magenta', label = 'Centroids')
plt.title('Clusters of customers (Hierarchical clustering)')
plt.xlabel('Annual income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1) 
plt.show()
```
